```python
Data={
    "Professors":["Dr. Mira","Mr. Shan Faaiz","Dr Sami ur rehman","Dr. Chandra Gupta"],
    "Subjects":["Computer Architechture","Mathematics","Python","Phylosophy and Dharma"],
    "ID":[123,124,125,126],
    "Universities":["Ardon","Madarsa","Punjab University","Kurukshetra University"],
    "Experience":[4,6,7,7],
    "Salary":[40000,20000,50000,80000],
    "Address":["Berlin","Potsdam","Shimla","Istanbul"]
}
df=pd.DataFrame(Data)
print(df)
print("Printing a single column form the data frames")
print(df["Professors"])
print("printing a rows using filters")
print(df[df["Salary"]>50000])
print("printing thee multiple columns")
print(df[["Professors","Salary","Address"]])
print("Printing the multiple rows")
print(df[(df["Salary"]>30000)&(df["Experience"])>3])
```
