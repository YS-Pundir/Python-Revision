```python
import pandas as pd
```

```python
data={
    "Names":["Ram","sham","ghansham","simran","Aditi","Raj","Simran"],
    "Age":[28,94,48,56,34,54,76],
    "Salary":[50000,30000,60000,70000,110000,660000,90000],
    "Performance":[85,90,78,92,88,95,89]
}

df=pd.DataFrame(data)
print(df)
#adding new column anywhere
df["Bonus"]=df["Salary"]*0.1
print("======New data frame======")
print(df)
# inserting a new column to the data frame at a precide location
df.insert(0,"Employee id",[1,2,3,4,5,6,7])
print("After inserting column on the precise location ")
print(df)
```
