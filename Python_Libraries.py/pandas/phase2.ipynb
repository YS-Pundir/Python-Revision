4. interpollution
    1. preserve data integrity
    2. smooth trend
    3. avoid data loss
    *Method to learn : linear  , polinomial,time

```python
data={
    "Names":["Ram",None,"ghansham","simran","Aditi","Raj","Simran"],
    "Age":[28,None,48,56,34,54,76],
    "Salary":[50000,None,60000,70000,110000,660000,90000],
    "Performance":[85,None,78,92,88,95,89]
}

df=pd.DataFrame(data)
df.interpolate(method="linear",axis=0,inplace=True)#axis=0 means row
df["Salary"]=df["Salary"].interpolate(method="linear")
df["Age"]=df["Age"].interpolate(method="linear")
df["Performance"]=df["Performance"].interpolate(method="linear")
df["Names"]=df["Names"].interpolate(method="linear")
print(df)
```

5. Sorting and Aggregration 
    1. soritng data in one column sort_values()
    df.sort_values(by="column naame",True/False,inplace=True)# true for ascending order or false fo rdescending order


```python
data={
    "Names":["Ram",None,"ghansham","simran","Aditi","Raj","Simran"],
    "Age":[28,20,48,56,34,54,76],
    "Salary":[50000,900000,60000,70000,110000,660000,90000],
    "Performance":[85,None,78,92,88,95,89]
}
df=pd.DataFrame(data)
df.sort_values(by=["Salary"],ascending=True,inplace=True)
print("Data frame with sorted by salary in ascending order")
print(df)
df.sort_values(by=["Age","Salary"],ascending=False,inplace=True)
print("Data frame sorted in descending order")
print(df)
```

6. Merging and Joining : 
* Merging : to combine the rows of more than two data frames
    1. Inner :
    2. Outer:
    3. left :
    4. Right:

```python
# pd.merge(df1,df2,on="Column_Name",how="Type of join")
costumers=pd.DataFrame({
    "CustomerID":[1,2,3],
    "Name":["Ramesh","Suresh","Jagdish"]
    
})

order=pd.DataFrame({
    "CustomerID":[1,2,4],
    "orderAmount":[250,350,400]
})
#inner join, only common value are merged
df_Merge=pd.merge(costumers,order,on="CustomerID",how="inner")
print("Inner Join")
print(df_Merge)
# outer join , all rows will be combined & the missng value will be filled by NaN
df_Merge2=pd.merge(costumers,order,on="CustomerID",how="outer")
print("Outer Join")
print(df_Merge2)
#left merge : will keep all the left values , and if some rigth sided values are missing then it will fill it with NaN
df_Merge=pd.merge(costumers,order,on="CustomerID",how="left")
print("Left Join")
print(df_Merge)
#Right merge =will keep all the right value and missing left values will be filled by NaN
df_Merge=pd.merge(costumers,order,on="CustomerID",how="right")
print("Right Join")
print(df_Merge)
```
