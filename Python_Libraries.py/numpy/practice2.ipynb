1. Broadcasting & Vectorization

    1. Subtract the mean of each column from a 2D matrix without using loops.
    2. Multiply each row of a matrix by a different scalar.

```python
import numpy as np
```

```python
#Subtract the mean of each column from a 2D matrix without using loops.
arr=np.arange(0,4).reshape(2,2)
arr= arr-arr[:,1].mean()-arr[:,0].mean()

print(arr)
#Multiply each row of a matrix by a different scalar.
arr[1]=arr[1]*2
arr[0]=arr[0]*5
print("the array after multipling it's row by scalar value:",arr)
```

2. Linear Algebra
    1. Compute :
        1. Matrix transpose
        2. Matrix inverse
        3. Determinant
    
    2. Solve a system of linear equations using NumPy.
    3. Verify the matrix multiplication rule: (AB)^T=(B^T)(A^T)

```python
print("The transpose of the matrix is :",arr.transpose())
print("The determinant of the matrix is : ",np.linalg.det(arr))
print("The inverse of the matrix :",np.linalg.inv(arr))
```

> The system of eqaution to solve :
1. ​2x+y−z=1
2. −3x−y+2z=−4
3. −2x+y+2z=−1​

```python
#let
a=np.array([[2,1,-1],[-3,-1,2],[-2,1,2]])
b=np.array([1,-4,-1])

ans=np.linalg.solve(a,b)
print("x=",ans[0])
print("y=",ans[1])
print("z=",ans[2])
```

```python
A=np.array([[2,1,-1],[-3,-1,2],[-2,1,2]])
B=np.array([[4,5,6],[8,9,7],[3,5,4]])
AB=np.dot(A,B).T
A_Trans=A.T
B_Trans=B.T
prod2=np.dot(B_Trans,A_Trans)
if AB.all()==prod2.all():
    print("Hence Proved!!")
else:
    print("Error")
```

3. Statistical & Aggregation Operations :
    1. Compute column-wise and row-wise:
        1. Mean
        2. Median
        3. Sum

    2. Find the index of:
        1. Maximum value in each row
        2. Minimum value in each column

```python
arr=np.array([[4,5,6],[8,9,7],[3,5,4]])
print("mean of first row of the matrix : ",arr[0].mean())
print("Mean fo first column : ",arr[:,0].mean())
print("Summition of first row of the matrix is ",arr[0].sum())
print("Summation of first column of the matrix : ",arr[:,0].sum())
```

```python
max=arr[0].max()
print("The maximum of first row of the matrix : ",np.where(arr==max))
print("The minimum value of first column : ",np.where(arr==arr[0].min()))
```
