1. Broadcasting & Vectorization

    1. Subtract the mean of each column from a 2D matrix without using loops.
    2. Multiply each row of a matrix by a different scalar.

```python
import numpy as np
```

```python
#Subtract the mean of each column from a 2D matrix without using loops.
arr=np.arange(0,4).reshape(2,2)
arr= arr-arr[:,1].mean()-arr[:,0].mean()

print(arr)
#Multiply each row of a matrix by a different scalar.
arr[1]=arr[1]*2
arr[0]=arr[0]*5
print("the array after multipling it's row by scalar value:",arr)
```
