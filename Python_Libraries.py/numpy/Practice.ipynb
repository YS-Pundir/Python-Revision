1. Create a 1D NumPy array of integers from 10 to 49.

2. Create a 3×3 matrix with values ranging from 0 to 8.

3. Create a 4×4 identity matrix.

4. Generate a 5×5 matrix filled with random floats in the range [0, 1).

5. Find the shape, dtype, ndim, and size of a given array.

```python
import numpy as np
```

```python
# creating  1D NumPy array of integers from 10 to 49
arr1=np.arange(10,50)
print("The first array is ",arr1)

#creating 3×3 matrix with values ranging from 0 to 8.
arr2=np.arange(0,9).reshape(3,3)
print("the second array is : ",arr2)
#Create a 4×4 identity matrix.
arr3=np.identity(4)
print("The third matrix is :",arr3)
# Creating a a 5×5 matrix filled with random floats in the range [0, 1).
arr4=np.random.random((5,5))
print("The fourth matrix is :",arr4)
#Finding the shape, dtype, ndim, and size of a given array.
print("The ahpe of the given array is : ",arr4.shape)
print("The datatype of the array is :",arr4.dtype)
print("The dimension of the array is :",arr4.ndim)
print("The size of the array is :",arr4.size)

```
