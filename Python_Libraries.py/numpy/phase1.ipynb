## Numpy array and basics

```python
import numpy as np
```

## creating arrays

```python
array1d=np.array([1,2,3])
print(array1d)

array2d=np.array([[1,2,3,4],[5,6,7,8]])
print(array2d)
```

```python
list=[1,2,3,4]
print("multiplication of 1d matrix : ",list*2)
print("multiplication of rd matrix : ",array1d*2)
```

### 2-> making diffrent arrays from scratch

```python
ones=np.ones((3,3))#this gonna make the matrices of one only , with provided order
print(ones)

zeros=np.zeros((2,3))
print("the ,atrix with full elements is :",zeros)

full=np.full((2,2),8)#first order of the matrix then the only elemnt of the matrix
print("the matrix with any elemnt is :", full)


random=np.random.random((2,3))# to have access to random amount
print("random array :",random)

sequence=np.arange(0,10,2)#array in sequence , start ,stop ,and the intervals
print("sequence array : ",sequence)




```

## vector , Matrix and Tensors
### 1. vector-> Vector is a one dimensional array , which means a single array.
### 2. Matrix-> Matrix is a array inside another array , something like 2 dimensions
### 3. Tenssors-> Tensors are the arrays which have more dimenisons then matrix
 


```python
vector=np.array([1,2,3,4])
print("vecotr:",vector)
matrix=np.array([[1,2,3],
                [4,5,6]])
print("matrix :", matrix)

tensors=np.array([[[1,2,3,2],[3,4,6,5]],
                  [[6,7,5,8],[9,5,6,3]]])
print("tensior : ", tensors)
```

### arrays properties

```python
arr=np.array([[1,2,3],
              [4,5,6]])
print("size : ",arr.size)# no of the elements

print("dimension :", arr.ndim)#the dimension of the array
print("shape",arr.shape)#the order of the matrix , in rows and column
print("DType : ",arr.dtype)# the data type
```

## array reshaping

```python
arr=np.arange(12)

print("original array :",arr)

reshaped=arr.reshape((3,4))
print("reshaped array : ",reshaped)

flattened=reshaped.flatten()
print("flate array : ",flattened)

#transpose
transpose=reshaped.T
print("the trasposed matrix is :",transpose)

```
