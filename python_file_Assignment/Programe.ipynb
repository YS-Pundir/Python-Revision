```python
my_list=[]
```

```python
try : 
    with open("numbers.txt","r") as file :
        print("File opened successfully")
        for line in file:
            my_list.append(int(line.strip()))
    print("Data loaded")    
except FileNotFoundError:
    print("File Not Found")
```

```python
total=len(my_list)
addition=sum(my_list)
average=sum(my_list)/total
print("calculation completed")
```

```python
with open("results.log","a") as f:
    f.write("File opened\n")
    f.write(f"read {total} numbers\n")
    f.write(f"Sum : {addition}\n")
    f.write(f"Average : {average}\n")
    f.write(f"Process completed \n")
```
