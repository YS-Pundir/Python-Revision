```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024
sales_data = np.array([[1, 150000, 180000, 220000, 250000],#pizza hut 
                    [2, 120000, 140000, 160000, 190000], # Beijing
                    [3, 200000, 230000, 260000, 300000] ,# Pizza Hub
                    [4, 180000, 210000, 240000, 270000], # Burger Point
                    [5, 160000, 185000, 205000, 230000]]# Chai Point
)

print("=====Zomato Sales Anylisis=====")
print("shape of the sales data : ",sales_data.shape)
print("sample data for first three restaurant : ",sales_data[0:3])
```

```python
#total sales per year
print(np.sum(sales_data[:,1:],axis=0))
#minimum sales
min_sales=np.min(sales_data[:,1:],axis=1)
print(min_sales)
#max sales per year 
max_sales=np.max(sales_data[:,1:],axis=0)
print("maximum sales per year : ",max_sales)
#avearge sales done by the restaurant 
avg_sales=np.mean(sales_data,axis=1)
print("maximum  sales done by per restaurant : ",avg_sales)

cumulative_sales=np.cumsum(sales_data[:,1:],axis=1)
print("the cumulative sales of the restauret is : ",cumulative_sales)

plt.plot(np.mean(cumulative_sales,axis=0))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Zomato Sales Anylisis",color="Red")
plt.show()

```
