```python
import pandas as pd
import plotly.express as px
# AS a first step of EDA we are going to analyse the data , by cheking the columns and datatypes 

# 1 : Understand the data 
data=pd.read_csv("titanic.csv")
data.info()
print("-"*50)
print("The other information about the data is : ")
data.describe()
```

```python
# 2. Inspect the data quality

print("Analysing the first five rows of data : ")
print(data.head())
print("Analysing the last rows of data : ")
print(data.tail())
print("analysing the null value in data: ")
print(data.isnull().sum())
```

```python
# 3 : Study the distribution : 

fig=px.histogram(
    data,
    x="Age",
    title="Age Distribution in Data"

)
fig.show()
```

```python
fig=px.histogram(
    data,
    x="Pclass",
    title="Age Distribution in Data"

)
fig.show()
```

```python
# 4 : detecting the outliers 

fig=px.box(
    data,
    y="Fare",
    title="Titanic Dataset's Outleirs"

)
fig.show()

fig=px.box(
    data,
    y="Age",
    title="Titanic Dataset's Outleirs"

)
fig.show()
```

```python
# 5 : Corelation measurement : 

        # This the matrix showing the corealtion with two varialbles with each others
corr = data.corr(numeric_only=True)
corr

fig=px.imshow(
    corr,
    text_auto=True,
    title="Variables' Corelation"
)
fig.show()

```
