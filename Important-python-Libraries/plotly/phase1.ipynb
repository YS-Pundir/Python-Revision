```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
data={
    "epchos":range(1,11),
    "Loss":[0.6,0.7,0.88,0.2,0.79,0.4,0.5,0.1,0.3,0.9]
    
    }

df=pd.DataFrame(data)
```

```python
fig=px.line(df,x=df["epchos"],y=df["Loss"],title="Training loss over time")
fig.show()
```

```python
data={
    "user":[3,5,7,2,8,9],
    "Perchase":[21,45,97,56,89,32]
}
df=pd.DataFrame(data)

fig=px.scatter(df,x=df["user"],y="Perchase",title="Perchase Analysis")
fig.show()
```

```python
df_cluster=pd.DataFrame({
    "x":np.random.rand(100),
    "y":np.random.rand(100),
    "group":np.random.choice(["A","B","C"],100)
})
print("The dataset is : ")
print(df_cluster)
## creat6ed a dataset and assigned a group to each row randomely

fig=px.scatter(df_cluster,x='x',y="y",color="group",title="Cluster Visulaisation",color_discrete_sequence=["#9b1fb4","#ff7f0e","#72ff0e"] )

fig.show()
```
