```python
df_cluster=pd.DataFrame({
    "x":np.random.rand(100),
    "y":np.random.rand(100),
    "group":np.random.choice(["A","B","C"],100)
})
print("The dataset is : ")
print(df_cluster)
## creat6ed a dataset and assigned a group to each row randomely

fig=px.scatter(df_cluster,x='x',y="y",color="group",title="Cluster Visulaisation",color_discrete_map={
    "A":"red",
    "B":"blue",
    "C":"Green"

} )
fig.write_image("PLot4.png")
fig.show()
```

```python
df_model=pd.DataFrame({
    "Model":["Model 1","Model 2","Model 3"],
    "Accuracy":[0.65,0.90,0.77]

})

fig=px.bar(df_model,x=df_model["Model"],y=df_model["Accuracy"],title="Modle accuracy comparison",color=["red","green","blue"])
fig.write_image("plot2.png")
fig.show()
```

```python
df=pd.DataFrame({
    "Age":[22,25,26,28,30,31,36,78,42,89,65,24,76]
})

fig=px.histogram(x=df["Age"],title="Age Distribution")
# fig=px.histogram(x=df["Age"],title="Age Distribution",nbins=4)
#usse nbins when we need to give control to the ssystem    on choosing the grouping 
# use xbins  when we want to hae control
fig.update_traces(
    xbins={
        "start":18,
        "end":50,
        "size":8
    
    }
)

fig.show()
```

```python
## Annotation
data={
    "epchos":range(1,11),
    "Loss":[0.6,0.7,0.88,0.2,0.79,0.4,0.5,0.1,0.3,0.9]
    
    }

df=pd.DataFrame(data)
fig=px.line(df,x=df["epchos"],y=df["Loss"],title="Training loss over time")
fig.add_annotation(
    x=5,
    y=0.79,
    text="Main Point",
    showarrow=True,
    arrowhead=6
)
fig.show()
```

```python
# usingh the klaeido library , saving the plot.img image
fig.write_image("plot.png")
```
