```python
data={
    "Names":["Ram",None,"ghansham","simran","Aditi","Raj","Simran"],
    "Age":[28,20,48,56,34,54,76],
    "Salary":[50000,900000,60000,70000,110000,660000,90000],
    "Performance":[85,None,78,92,88,95,89]
}

df=pd.DataFrame(data)

df["Performance"]=df["Performance"].interpolate(method="linear")
print(df)
print(df.isnull())
df.loc[1,"Names"]="Yuvraj"
print(df)
df.index=["a","b","c","d","e","f","g"]
print(df)
df.loc["a","Names"]="Yuvraj"
print(df)

```
