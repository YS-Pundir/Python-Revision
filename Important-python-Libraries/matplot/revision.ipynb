```python
import pandas as pd 
import matplotlib.pyplot as plt

data={
    "Day":[1,2,3,4,5],
    "Users":[120,150,170,160,180],
    "Purchase":[30,35,40,38,45]
}

df=pd.DataFrame(data)
print(df)
```

```python
plt.plot(df["Day"],df["Users"])
plt.xlabel("Day")
plt.ylabel("No of Users")
plt.title("Random Shit")
```

```python
plt.plot(df['Day'],df["Purchase"])
plt.xlabel("Days")
plt.ylabel("No of Purchases")
plt.title("Another Bar chart")
```

```python
plt.scatter(df["Users"],df["Purchase"])
plt.xlabel("Users")
plt.ylabel("Amount of perchases")
plt.title("Users V/S Perchases")
```

```python
plt.figure()
plt.plot(df['Day'],df["Purchase"])
plt.xlabel("Days")
plt.ylabel("No of Purchases")
plt.title("Another Bar chart")

plt.figure()
plt.scatter(df["Users"],df["Purchase"])
plt.xlabel("Users")
plt.ylabel("Amount of perchases")
plt.title("Users V/S Perchases")
```

```python
plt.figure(figsize=[10,4])
plt.subplot(1,2,1)
plt.plot(df["Day"],df["Users"],linestyle=":",color="Green",marker="s",markerfacecolor="black",linewidth=1)# for circle mark : "o","s":square,"t":Trigangle
plt.xlabel("Days",color="blue")
plt.ylabel("No of users",color="Green")
plt.title("Another Bar chart",color="Red")
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(df["Users"],df["Purchase"],marker="*",color="Green")
plt.xlabel("Users",color="blue")
plt.ylabel("Amount of perchases",color="Green")
plt.title("Users V/S Perchases",color="red")

plt.grid(True)
```
