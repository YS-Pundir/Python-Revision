```python
parameters={
    "q":"Agentic",
    "sort":"stars",
    "per_page":4,
    "order":"desc"

}

response=requests.get("https://api.github.com/search/repositories",params=parameters)
data=response.json()
print(data["items"])
print("-"*50)
for i in data["items"]:
    print(i["full_name"])
```
