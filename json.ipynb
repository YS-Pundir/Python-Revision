* How to convert dictionary to JSON 


```python
import json
data={
    "Name":"Yuvraj SIngh",
    "Roll No":231359,
    "Class":"Bsc ist sem"
}
with open("config.json","w") as f:
    json.dump(data,f)

json.dumps(data)
```

> now how to convert json data to dictionary


```python
reponse='''{
"id":"2917391",
"Name":"Yuvraj Sngh",
"course":"software engineering"

}'''
# to convert from string usse loads
parsed=json.loads(reponse)
print(parsed)
#to convert from file to dictionary , then use load
with open("config.json","r") as f:
    data=json.load(f)
print(data)
```
