5. ### logerthmic time :

```python
#let us condider a programe for it
n=10
while (n>=1):
    print("Hello World")  # let the loop will run k times 
                           # The coplexity would be O(k) , and we will be finding the value of k first
    n=n/2
```

> The condition at  which everytime this loop will run : n/2^1 , n/2^2 , n/n^3 .....n/2^k . 
* As the condition of while loop is n>=1  , so the n/2^k=1.
* => n=2^k
* =>log(n)=k(log(2)) => k = log(n)/log(2) ==log(n)(neglecting the non dominating factor)
* Hence Our time complexity of the program would be  O(log(n))
